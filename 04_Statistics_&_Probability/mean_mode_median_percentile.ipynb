```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "name":["Rob","Rafiq","Nina","Mina","Tina","Lina","Bil Gates"],
    "income":[5000,6000,4000,7500,8000,7000,20000000]
})
df
```

```python
df.describe()
```

```python
df.income.quantile(0.25)
```

```python
df.income.quantile(0.25, interpolation='higher')
```

```python
df.income.quantile(0.25, interpolation='lower')
```

```python
df.income.quantile(0.75)
```

```python
df.income.quantile(1)
```

```python
#removing outliers using quantile

df_percentile=df.income.quantile(0.99)
df_percentile
```

```python
df_no_outliers = df[df.income<=df_percentile]
df_no_outliers
```

```python
#making a row NaN

df['income'][3]=np.NaN
df
```

```python
df.income.mean()
```

```python
df.income.median()
```

```python
df.income.mode()
```

```python
#filling the missing value with the median

df_new=df.fillna(df.income.median())
df_new
```
