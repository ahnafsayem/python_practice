```python
import pandas as pd
```

```python
df = pd.read_csv('/content/weather_by_cities.csv')
df
```

```python
#created a dataframe by groupby in cities

g = df.groupby('city')
g
```

```python
for city, data in g:
  print("City Name: ", city)
  print('Data: ',data)
  print('\n')
```

```python
for city, data in g:
  print("City Name: ", city)
  print('Data: ',data.temperature.max())
  print('\n')
```

```python
def grouper(df,idx,col):
  if 80<= df[col].loc[idx] <=90:
    return '80-90'
  elif 50<= df[col].loc[idx] <=60:
    return '50-60'
  else:
    return 'others'
```

```python
n=df.groupby(lambda idx: grouper(df, idx, 'temperature'))
n
```

```python
for key, data in n:
  print("Key: ",key)
  print("Data: ",data)
```
