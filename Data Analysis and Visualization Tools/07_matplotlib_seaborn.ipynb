```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sales=pd.read_excel('/content/linechart.xlsx')
df_sales
```

```python
# figure size
plt.figure(figsize=(12,4))

# plotting
plt.plot(df_sales['Quarter'], df_sales['Fridge'], color='red', label='Fridge')
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'], color='green', label='Dishwasher')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'], color='blue', label='Washing Machine')

# decorating
plt.title("Product Sales")
plt.xlabel("Quarter")
plt.ylabel("Revenue in $")
plt.legend()

# display
plt.show()
```

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct="%1.2f%%",
        explode=(0.08,0,0),shadow=True, startangle=140)

plt.show()
```

```python
df_sales.plot(kind="bar",x="Quarter")
plt.xticks(rotation=45)

plt.show()
```
