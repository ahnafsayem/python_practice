```python
import pandas as pd
```

```python
bd_temp = pd.DataFrame({
    "City":["Dhaka","Rajshahi","Chittagong","Khulna","Sylthet"],
    'Temperature':[32,40,30,41,28],
    'Humidity':[70,80,65,75,55]
})

bd_temp
```

```python
us_temp = pd.DataFrame({
    "City":["NY","SEA","TX","PYV","CA"],
    'Temperature':[20,12,30,8,22],
    'Humidity':[40,10,45,20,15]
})

us_temp
```

```python
df = pd.concat([bd_temp,us_temp])
df
```

```python
df = pd.concat([bd_temp,us_temp], keys=['BD','US'])
df
```

```python
df.loc['BD']
```

```python
df.loc['US']
```

```python
df1 = pd.DataFrame({
    "City":["Dhaka","Rajshahi","Chittagong","Khulna","Sylthet"],
    'Temperature':[32,40,30,41,28]
})
df1
```

```python
df2 = pd.DataFrame({
    "City":["Dhaka","Rajshahi","Chittagong","Khulna","Sylthet"],
    'Humidity':[70,80,65,75,55]
})
df2
```

```python
df12_merged=pd.merge(df1,df2, on='City')
df12_merged
```

```python
us1 = pd.DataFrame({
    "City":["NY","SEA","TX"],
    'Temperature':[20,12,30]
})

us1
```

```python
us2 = pd.DataFrame({
    "City":["NY","TX","PYV","CA"],
    'Humidity':[40,45,20,15]
})

us2
```

```python
us12_merged=pd.merge(us1,us2, on='City')
us12_merged
```

```python
us12_merged=pd.merge(us1,us2, on='City',how='inner')
us12_merged
```

```python
us12_merged=pd.merge(us1,us2, on='City',how='left')
us12_merged
```

```python
us12_merged=pd.merge(us1,us2, on='City',how='right')
us12_merged
```

```python
us12_merged=pd.merge(us1,us2, on='City',how='outer')
us12_merged
```
