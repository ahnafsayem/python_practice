```python
#Null value management

df = pd.read_csv("/content/stock_data.csv", header=1, na_values=['not available',-1,'n.a.'])
df.head()
```

```python
"""
price to earning ratio, (pr = price/eps) , eps means earnings per share
"""

df["pe"]=df['price']/df['eps']
df
```

```python
#calling multiple sheets at once from excel file

df_movies = pd.read_excel("/content/movies_db.xlsx","movies")
df_movies.head(5)
```

```python
df_financials = pd.read_excel("/content/movies_db.xlsx","financials")
df_financials.head(5)
```

```python
#converting currencies
def standardize_currency(curr):
  if curr == '$$' or curr=="Dollars":
    return 'USD'
  return curr

df_financials = pd.read_excel("/content/movies_db.xlsx","financials", converters={
    'currency':standardize_currency
})
df_financials.head(5)
```

```python
df_merged = pd.merge(df_movies,df_financials,on="movie_id")
df_merged.head(5)
```

```python
df_merged.to_excel("movies_merged.xlsx",sheet_name="merged",index=False)
df_merged.head(5)
```

```python
df_stocks = pd.DataFrame({
    'tickers':['GOOGlE','WMT','MSFT'],
    'price':[800,50,60],
    'pe':[14.55,40.15,50.15],
    'eps':[20,22,30],
})
df_stocks
```

```python
df_weather = pd.DataFrame({
    'Date':['01 Jan','01 May','01 Oct'],
    'Weather':['Cold','Sunny','Rainy'],
    'Temp(Cel.)':[10,40,26],
})
df_weather
```

```python
with pd.ExcelWriter('Stocks_Weather.xlsx') as writer:
  df_stocks.to_excel(writer, sheet_name='stocks')
  df_weather.to_excel(writer, sheet_name='weather')
```
