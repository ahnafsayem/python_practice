```python
#changing the data type of the data

df = pd.read_csv('/content/weather_data.csv', parse_dates=['day'])
df.head(5)
```

```python
type(df.day[0])
```

```python
#setting day as the index

df.set_index('day', inplace=True)
df.head(5)
```

```python
df.fillna(0) #to change the whole daatset permanently, add inplace=true
```

```python
df.fillna(method='ffill') #ffill=forward fill, bfill=backward fill
```

```python
df.fillna({
    'temperature':df.temperature.mean(),
    'windspeed':df.windspeed.mean(),
    'event':'No Event'}) #inplace=True)

df
```

```python
df
```

```python
df.interpolate() #gives a linear data like 32 then 30 and 28
```

```python
df.dropna()
```

```python
df.dropna(how='all') #drops the row that has all NaN value
```

```python
df.dropna(thresh=2) #drops the row that has min 2 NaN value
```
