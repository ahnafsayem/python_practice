Main goal for working in this notebook is to handle the outlires of this given dataset.

```python
#Importing necessary Libraries

import numpy as np
import pandas as pd
```

```python
#Load Data

df=pd.read_csv("/kaggle/input/new-york-city-airbnb-open-data/AB_NYC_2019.csv")
df.head(5)
```

```python
#Basic Information of the dataset

df.info()
```

```python
#Describe the dataset

df.price.describe()
```

```python
#sorting the dataset in the ascending order

df_sorted = df.sort_values(by="price")
df_sorted.head(2)
```

```python
df_sorted = df.sort_values(by="price")
df_sorted.tail(2)
```

```python
#Filtered out data range whithout the outliers

min_threshold,max_threshold = df.price.quantile([0.01,0.999])
min_threshold,max_threshold
```

```python
#new dataset without outlires

df_2 = df[(df.price>min_threshold)&(df.price<max_threshold)]
df_2
```

```python
df_2.describe()
```
